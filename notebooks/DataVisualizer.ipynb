## Imports

```
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

## Data paths

```
train_data_path = "/content/drive/My Drive/documents/projects/DeCaf/data/new/train_data/processed/combined.csv"
test_data_path = "/content/drive/My Drive/documents/projects/DeCaf/data/new/test_data/processed/test_data.csv"
validation_data_path = "/content/drive/My Drive/documents/projects/DeCaf/data/new/validation_data/processed/validation_data.csv"
```

## Word Cloud

```
def create_word_cloud(data):
  design_data_string = data[data.label == 'design'].text.str.cat(sep='\n')
  general_data_string = data[data.label == 'general'].text.str.cat(sep='\n')
  fig, (ax_design, ax_general) = plt.subplots(1, 2)
  fig.set_figheight(12)
  fig.set_figwidth(25)

  design_word_cloud = WordCloud(max_font_size=50, background_color='white', collocations=False).generate(design_data_string)
  general_word_cloud = WordCloud(max_font_size=50, background_color='white', collocations=False).generate(general_data_string)

  ax_design.imshow(design_word_cloud)
  ax_general.imshow(general_word_cloud)

  ax_design.set_title("Design Words")
  ax_general.set_title("General Words")

  plt.show()
```

## Visualize train data

```
train_data = pd.read_csv(train_data_path)
```

### Word Cloud

```
create_word_cloud(train_data)
```

## Visualize test data

```
test_data = pd.read_csv(test_data_path)
```

### Word Cloud

```
create_word_cloud(test_data)
```

## Visualize validation data

```
validation_data = pd.read_csv(validation_data_path)
```

### Word Cloud

```
create_word_cloud(validation_data)
```
